```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential 
from keras.layers import Dense
from keras.optimizers import SGD
```

```python

# Cargar el conjunto de datos
df = pd.read_csv('F1drivers_Dataset.csv')

# Extraer las columnas 'Pole_Positions' y 'Race_Wins' y convertirlas a arrays de numpy
pole_positions = df['Pole_Positions'].to_numpy()
race_wins = df['Race_Wins'].to_numpy()

# Imprimir los datos extraídos
print("Posiciones de Pole:", pole_positions)
print("Victorias de Carreras:", race_wins)
```

```python
# Se crea elp plot y sus subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Grafico de dispersión para Pole Positions
axs[0, 0].scatter(range(len(pole_positions)), pole_positions)
axs[0, 0].set_title('Grafico de dispersión Pole Positions')

# Grafico de disepercion para Race Wins
axs[0, 1].scatter(range(len(race_wins)), race_wins, color='orange')
axs[0, 1].set_title('Grafico de dispersion  Race Wins')

# Grafico de dispersion para Pole Positions vs Race Wins
axs[1, 0].scatter(pole_positions, race_wins, color='violet')
axs[1, 0].set_xlabel('Pole Positions')
axs[1, 0].set_ylabel('Race Wins')
axs[1, 0].set_title('Pole Positions vs Race Wins Scatter Plot')

# Grafico de dispersion  en conjunto
axs[1, 1].scatter(range(len(pole_positions)), pole_positions, label='Pole Positions')
axs[1, 1].scatter(range(len(race_wins)), race_wins, color='orange', label='Race Wins')
axs[1, 1].set_title('Pole Positions and Race Wins ')
axs[1, 1].legend()


plt.tight_layout()
plt.show()
```

```python
# Definir los datos de entrenamiento
data_train_x = pole_positions
data_train_y = race_wins

# Ccrea un moedelo secuencial de enrtrenamiento
model = Sequential()

# Añade una capa densa con una sola neurona y una entrada de una sola dimensión
model.add(Dense(1, input_shape=(1,)))

# Ccompila el modelo con el optimizador SGD y la función de pérdida de error cuadrático medio
model.compile(loss='mean_squared_error', optimizer=SGD(learning_rate=0.01))

```

```python
#historial de entrenamiento del modelo  y el conjunto de datos de entrenamiento y empieza el entrenamiento
history = model.fit(data_train_x, data_train_y, epochs=2000, validation_split=0.2, batch_size=3, verbose=1)
```

```python
# Extraer las pérdidas del historial de entrenamiento
losses = history.history['loss']
print(losses)

#graficar el progreso de la pérdida durante el entrenamiento
plt.plot(losses)
plt.title('Model Loss Progression During Training/Validation')
plt.ylabel('Training Loss')
plt.xlabel('Epoch Number')
plt.show()

```

```python
# Generar los datos de prueba para precidir las victorias de carreras con base en l las ´pole positions
test_race_wins = np.random.randint(low=0, high=100, size=10)
test_pole_positions = np.random.randint(low=0, high=100, size=10).reshape((10, 1, 1))
pole_positions_pred = model.predict(test_pole_positions)

# imprime los datos de prueba y las predicciones
print("Test Pole Positions:", test_pole_positions)

# calcula e imprime el error cuadrático medio
mse = np.mean((test_race_wins - pole_positions_pred)**2)
print("Mean Squared Error:", mse)
```

```python
plt.plot(test_pole_positions.flatten(), test_race_wins, "r", label="Original Data")
plt.plot(test_pole_positions.flatten(), pole_positions_pred.flatten(), "b", label="Predicted Data")
plt.legend()
plt.xlabel('Pole Positions')
plt.ylabel('Race wins')
plt.title('Comparison between original and predicted data')
plt.show()
```
