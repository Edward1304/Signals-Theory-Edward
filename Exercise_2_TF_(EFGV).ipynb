# Ejercicio 1: Sistema de Biblioteca 

Crea una clase Libro que contenga atributos como título, autor, año de publicación y estado (disponible o prestado). Luego, desarrolla una clase Biblioteca que gestione una colección de objetos Libro. Esta clase debe incluir métodos para: 

Añadir un nuevo libro al catálogo. 

Prestar un libro a un usuario (cambiar el estado a prestado). 

Devolver un libro (cambiar el estado a disponible). 

Buscar libros por autor o título. 

Mostrar los libros disponibles para préstamo. 



```python

class book:
    """
    Clase "book" , una clase que represnta un libro  para una bibliotes.

    Atributos:
    * title (str): Entrada de texto que indica el titulo del libro.
    * author (str): Entradad de texto que  es el nombre del auutor.
    * year_publication (int): una entrada numerica entera que va ser el año de publicacioon.
    * status (str): Entrada de texto que indica que si el libro esta prestado o esta disponible".
    """
    
    def __init__(self, title, author, year_publication):
        self.title = title
        self.author = author
        self.year_publication = year_publication
        self.status = "disponible"



class library:
    """
    Clase "library", es una clase que corresponde a  la bilioteca.

    Atributos:
    - catalogue (list): recibe una lista  lo cual va corresponder a la informacion de los libros.
    """

    def __init__(self):
    
        self.catalogue = []

    def add_book(self, book):
        """
        Añadir el libro al catalogo de la biblioteca, teniendo como parametro de entrada el libro.

        """

    def lend_book(self, title):
        """
        Prestar un libro de la   biblioteca.

        Parametros:
        - title (str): Entrada de texto , el cual se debe ingrear el titulo del libro para prestarlo.

        Retorna:
        - str: devuelve un mensaje de texto  que  le va indicar  de que el libro  se ha prestado o no esta
        disponible o el libro no se ha encontrado.
        """

        for book in self.catalogue:
            if book.title == title:
                if book.status == "disponible":
                    book.status = "prestado"
                    return f"El libro '{title}' ha sido prestado."
                else:
                    return f"El book '{title}' no está disponible para préstamo."
        return f"No se encontró el book '{title}' en la biblioteca."

    def back_Book(self, title):
        """
        Devolver  un libro a la biblioteca.

        Parametros:
        - title (str): ingresa en texto el titutlo del libro que quiere devolver.

        Retorna:
        - str: Devulve el mensaje de texto que indica  el libro ha sido devuelto o el libro esta prestado o el libro no se ha encontrado.
        """
        for book in self.catalogue:
            if book.title == title:
                if book.status == "prestado":
                    book.status = "disponible"
                    return f"El libro '{title}' ha sido devuelto."
                else:
                    return f"El libro '{title}' esta prestado."
        return f"No se encontró el libro '{title}' en la biblioteca."

    def search_books(self, author=None, title=None):
        """
        Busca los libros en el catalogo de la biblioteca .

        Parametros:
        - author (str): entrada de texto para buscar el nombre  del autor , teniendo que sea opcional el "none".
        - title (str): entrada de texto para buscar el titulo del lirbo .

        Retorna:
        - list: Devuelve una lista de libros que se han encontrado en el catalogo de la biblioteca.
        """
        found_books = []
        for book in self.catalogue:
            if author and book.author.lower() == author.lower():
                found_books.append(book)
            elif title and book.title.lower() == title.lower():
                found_books.append(book)
        return found_books

    def show_available_books(self):
        """
        Mostrar los libros que estan disponible.

        Retorna:
        - list: Devuelve una lista de los libros que estan disponibles.
        """
        books_available = []
        for book in self.catalogue:
            if book.status == "disponible":
                books_available.append(book)
        return books_available
```

# Ejercicio 2: Sistema de Reservas de Hotel 

- Define una clase Habitacion con atributos como número de habitación, tipo (simple, doble, suite), precio por noche y estado (ocupada o disponible). Luego, crea una clase Hotel que administre varias habitaciones. La clase Hotel debe ofrecer funcionalidades para:

* Registrar una nueva habitación.\
* Reservar una habitación, cambiando su estado a ocupado.\
* Liberar una habitación después de la estancia, cambiando su estado a disponible.
*Calcular el costo de la estancia de un cliente.\
* Mostrar todas las habitaciones disponibles. 


```python
class Room:
    """
    Clase de habitaciones de un holte.

    Atributos:
    - room_number (int): entrada de valor numerica entera para indicar el  numero de la habitacion.
    - room_type (str): Entrada de texto para indicar el tipo que es la habitacion.
    - value_night (float): Entreda numerica flotante para indicar el precio de la noche ne la habitacion .
    - status (str): Entrad de texto para indicar el estado en la que se encuentra la habitacion "Disponible" 0 "ocupada".

    Metodos:
    - reserve(): Reservar si la habitacion esta disponible.
    - release(): Liberare su la habitacion esta ocupda.
    """

    def __init__(self, room_number, room_type, value_night):
        self.room_number = room_number
        self.room_type = room_type
        self.value_night = value_night
        self.status = "Disponible"

    def reserve(self):
        if self.status == "Disponible":
            self.status = "ocupada"
            return f"Habitacion {self.room_number} ha sido reservada."
        else:
            return f"Habitacion {self.room_number} está ocupada."

    def release(self):
        if self.status == "ocupada":
            self.status = "Disponible"
            return f"Habitacion {self.room_number} está liberada."
        else:
            return f"Habitacion {self.room_number} está disponible."


class Hotel:
    """
    Clase  Hotel  que represtan  un con habitaciones que se pueden registrar, reservar, liberar y verificar disponibilidad.
    """

    def __init__(self):
        self.rooms = []

    def register_room(self, room):
        self.rooms.append(room)
    
    def reserve_room(self, room_number):
        """
        Reserva una habitacion con el numero de habitacion dado.
        
        Parametros:
        - room_number (int): El numero de la habitacion.
        
        Retorna:
        - str:Devuelve el mensaje del estado de la reserva.
        """
        for room in self.rooms:
            if room.room_number == room_number:
                return room.reserve()
        return f"Habitacion {room_number}  no existe."

    def release_room(self, room_number):
        """
        Libera una habitacion con el numero dado.
        
        Parametros:
        - room_number (int): el numero de la habitacion que se quiere dejar libre.
        
        Returns:
        - str: Devuelve el mensaje del estado  si queda libre o no.
        """
        for room in self.rooms:
            if room.room_number == room_number:
                return room.release()
        return f"habitacion {room_number} no existe."

    def calculate_cost(self, room_number, num_nights):
        """
        Calcula el costo total  por permanecer en una habitacion durante un numero determinado de noches.
        
        Parametros:
        - room_number (int): El numero de habitacion para calcular el costo.
        - num_nights (int): El numero de noches que estuvo en el hotel.
        
        Returns:
        - float: Devuelve el valor total  en la estadia en el.
        """
        for room in self.rooms:
            if room.room_number == room_number:
                return room.value_night * num_nights
        return f"Habitacion {room_number} no existe"

    def show_available_rooms(self):
        """
        Muesta una lista de la habitaciones libres en el hotel.
        
        Returns:
        - list: Devuelve una lista que muestras las habitaciones que estan disponibles.
        """
        available_rooms = []
        for room in self.rooms:
            if room.status == "Disponible":
                available_rooms.append(room)
        return available_rooms
    


```

##  Prueba  de Funcionamiento  Ejercicio 2

```python
# Crear algunas habitaciones
Room1 = Room(101, "Individual", 50)
Room2 = Room(102, "Doble", 80)
Room3 = Room(103, "Suite", 120)

# Crear un hotel
hotel = Hotel()

# Registrar las habitaciones en el hotel
hotel.register_room(Room1)
hotel.register_room(Room2)
hotel.register_room(Room3)

# Mostrar todas las habitaciones disponibles en el hotel
print("Habitaciones disponibles en el hotel:")
print(hotel.show_available_rooms())

# Reservar una habitación
print(hotel.reserve_room(101))

# Intentar reservar una habitación que ya está ocupada
print(hotel.reserve_room(101))

# Mostrar todas las habitaciones disponibles en el hotel cuando ya se ha reservado una
print("Habitaciones disponibles en el hotel después de reservar una:")
print(hotel.show_available_rooms())

# Liberar una habitación
print(hotel.release_room(101))

# Intentar liberar una habitación que ya está disponible
print(hotel.release_room(101))

# Calcular el costo de una reserva
print("Costo de reserva de habitación 102 por 3 noches:", hotel.calculate_cost(102, 3))

# Mostrar todas las habitaciones disponibles en el hotel después de liberar una
print("Habitaciones disponibles en el hotel después de liberar una:")
print(hotel.show_available_rooms())
```

# Ejercicio 3: Sistema de Gestión de Empleados

Implementa una clase Empleado con atributos como nombre, identificación, puesto y salario. Posteriormente, desarrolla una clase Departamento que contenga una lista de empleados. Esta clase debe ser capaz de:

+ Añadir un nuevo empleado. 
+ Eliminar un empleado por su identificación.
+ Aumentar el salario de un empleado por un porcentaje dado.
+ Mostrar la información de todos los empleados. 
+ Calcular el gasto total en salarios dentro del departamento.


```python
class Employee:
    """
    Clase empleado que tiene toda la informacion de cada empleado,
    Atributos:
    - name (str): Entrada de texto que indica el nombre del empleado.
    - id (int): Entrada numerica entera que indica la identificacion del empleado.
    - job (str): Entrada de texto que indica el puesto de trabajo del empleado.
    - salary (float): Entrada numerica flotante que indica el salario del empleado.

    """

    def __init__(self, name, id, job, salary):
        self.name = name
        self.id = id
        self.job = job
        self.salary = salary

class Department:
    """
    Clase Departamento de la empresa.
    
    Atributos:
    - staff (list): Es una lista de todo el conjunto de trabajadores que hay  en la empresa(departamento).
    
    Metodos:
    - add_employee(employee): Agrega un empleado al departamento.
    - remove_employee(id): Elimina a un empleado del departamento y lo hace por su Identificador(ID).
    - increase_salary(id, porcent): incremeta el salario del empleado en  base de si ID y un porcentaje dado.
    - show_staff(): mustra la informacion de todod los empleados que estan en el departamento.
    - calculate_salary_spend(): Calcucula el gasto total en salario en el departamento.
    """
    def __init__(self):
        self.staff = []

    def add_employee(self, employee):
        self.staff.append(employee)

    def remove_employee(self, id):
        for employee in self.staff:
            if employee.id == id:
                self.staff.remove(employee)
                return f"Empleado con identificación {id} eliminado."
        return f"No se encontró ningún empleado con identificación {id}."

    def increase_salary(self, id, porcent):
        for employee in self.staff:
            if employee.id == id:
                employee.salary *= (1 + porcent / 100)
                return f"Salario del empleado con identificación {id} aumentado en {porcent}%."
        return f"No se encontró ningún empleado con identificación {id}."

    def show_staff(self):
        for employee in self.staff:
            print(f"Nombre: {employee.name}, Identificación: {employee.id}, Puesto: {employee.job}, salary: {employee.salary}")

    def calculate_salary_spend(self):
        total_salary = sum(employee.salary for employee in self.staff)
        return f"Gasto total en salarios: {total_salary}"
```

# Ejercicio 4: Tienda en Línea
Crea una clase Producto con atributos tales como identificador, nombre, precio y cantidad en stock. Después, diseña una clase CarritoDeCompras que permita a los usuarios agregar o eliminar productos, y calcular el total de su compra. La clase CarritoDeCompras debe ofrecer métodos para:

+ Añadir un producto al carrito.
+ Eliminar un producto del carrito.
+ Mostrar los productos en el carrito.
+ Calcular el total de la compra.
+ Aplicar descuentos a productos específicos o al total de la compra.


```python
class Product:
    """
    Clase de producto.

    Atributos:
        id_product (int): Identificador unico de cada producto.
        name (str): El nombre del producto.
        value (float): Precio del  producto.
        quantity_stock (int): La cantidad de cada producto quee hay en estoy.
    """

    def __init__(self, id_product, name, value, quantity_stock):
        self.id_product = id_product
        self.name = name
        self.value = value
        self.quantity_stock = quantity_stock

class ShoppingCart:
    """
    Clase de carrito de compras  que va tener los  productos elegidos.

    Atributos:
    - products: Una lista de productos que estan en el carrito de compras.

    Metodos:
    - add_product: Agrega el producto  al carrito de compras.
    - delete_product: Elimina el producto del carrito de compras.
    - show_products: Muestas los producto y los detalles de estos en el carrito de conpras.
    - clclt_total_shop: Calcula el valor tocas de los producto en el carrito de compras.
    - apply_product_discount: Aplica el descuento a cada producto en especifico a cada producto.
    - apply_total_discount: Aplica el descuento a todos los productos en el carrito de compras.
    """
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def delete_product(self, id_product):
        for product in self.products:
            if product.id_product == id_product:
                self.products.remove(product)
                return f"Producto con identificador {id_product} eliminado del carrito."
        return f"No se encontró ningún producto con identificador {id_product} en el carrito."

    def show_products(self):
        for product in self.products:
            print(f"id_product: {product.id_product}, Nombre: {product.name}, Precio: {product.value}, Cantidad en stock: {product.quantity_stock}")

    def clclt_total_shop(self):
        total = sum(product.value for product in self.products)
        return f"Total de la compra: {total}"

    def apply_product_discount(self, id_product, discount):
        for product in self.products:
            if product.id_product == id_product:
                product.value -= product.value * (discount / 100)
                return f"Descuento de {discount}% aplicado al product con identificador {id_product}."
        return f"No se encontró ningún producto con identificador {id_product} en el carrito."

    def apply_total_discount(self, discount):
        for product in self.products:
            product.value -= product.value * (discount / 100)
        return f"Descuento de {discount}% aplicado a todos los productos en el carrito."
```
