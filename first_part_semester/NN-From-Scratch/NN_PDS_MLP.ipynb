```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

'''
Se utiliza un regresor de perceptrn multicapa (MLP) para predecir la densidad espectral de potencia (PSD) de una seal.
'''

'''
datos creado que  son de una seal sinusoidal con ruido aadido
'''
# Generar datos 
x = np.linspace(0, 10, 1000)
y = np.sin(x) + np.random.normal(0, 0.1, 1000)

'''
Se crea unmodelo MLPRegressor con dos capas ocultas de 10 neuronas cada una, 
utilizando la funcin de activacin ReLU y el solucionador Adam.
'''

# Crear el modelo de la red neuronal
model = MLPRegressor(hidden_layer_sizes=(10, 10), activation='relu', solver='adam', max_iter=1000)
# Entrenar el modelo
model.fit(x.reshape(-1, 1), y)


'''
Se Predice la PSD de una seal de prueba utilizando el modelo entrenado.
'''
# Predecir el PSD de una seal
x_test = np.linspace(0, 10, 1000)
y_pred = model.predict(x_test.reshape(-1, 1))

# Visualizar los resultados
plt.plot(x, y, label='Seal real')
plt.plot(x_test, y_pred, label='Predicha')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
