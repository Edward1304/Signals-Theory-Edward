```python
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Create a sequential model
model = Sequential()

# Add the input layer
model.add(Dense(units=64, activation='relu', input_dim=10))

# Add the first hidden layer
model.add(Dense(units=32, activation='relu'))

# Add the second hidden layer
model.add(Dense(units=16, activation='relu'))

# Add the output layer
model.add(Dense(units=1, activation='sigmoid'))

# Compile the model
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

# Print the model summary
model.summary()

# Generar datos de entrenamiento aleatorios
X_train = np.random.rand(100, 10)  # Ejemplo: 100 muestras con 10 características cada una
y_train = np.random.randint(0, 2, size=(100, 1))  # Ejemplo: 100 etiquetas binarias (0 o 1)

# Utiliza los datos de entrenamiento generados en el código existente

model.fit(X_train, y_train, epochs=10, batch_size=32)
model.fit(X_train, y_train, epochs=10, batch_size=32)
import matplotlib.pyplot as plt

# Plot the values of X_train
plt.figure(figsize=(10, 5))
plt.plot(X_train)
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title('X_train Data')
plt.show()




# END: FILEPATH

```

```python
# Get the training history
history = model.fit(X_train, y_train, epochs=10, batch_size=32)

# Plot the error
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Training Error')
plt.show()
```
